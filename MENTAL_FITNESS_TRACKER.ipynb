IMPLEMENTATION

```python
#normalising the labels
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
for i in df.columns:
  if df[i].dtype == 'object':
    df[i] = l.fit_transform(df[i])
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
x = df.drop('Eating_Disorder', axis=1)
y = df['Eating_Disorder']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.20, random_state=3)
```

1. LINEAR REGRESSION

```python
#training the data
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
#evaluating the data
from sklearn.metrics import mean_squared_error, r2_score
ytrain_pred = lr.predict(xtrain)
ytest_pred = lr.predict(xtest)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
r2 = r2_score(ytrain, ytrain_pred)

print("MSE : {}".format(mse))
print("RMSE : {}".format(rmse))
print("R2 : {}".format(r2))
```

```python
#scatter plot
plt.scatter(ytest, ytest_pred, color='blue')
plt.xlabel("Labelled Data")
plt.ylabel("Predicted data")
plt.show()
```

2. RANDOM FOREST REGRESSION

```python
#training the data
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain,ytrain)
```

```python
#evaluating the data
from sklearn.metrics import mean_squared_error, r2_score
ytrain_pred = rf.predict(xtrain)
ytest_pred = rf.predict(xtest)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
r2 = r2_score(ytrain, ytrain_pred)

print("MSE : {}".format(mse))
print("RMSE : {}".format(rmse))
print("R2 : {}".format(r2))
```

```python
#scatter plot
plt.scatter(ytest, ytest_pred, color='blue')
plt.xlabel("Labelled Data")
plt.ylabel("Predicted data")
plt.show()
```
